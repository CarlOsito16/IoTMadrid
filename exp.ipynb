```python
def assign_pvalue_star(p_value):
    if p_value <= 0.01:
        result = "***"
    elif p_value <= 0.05:
        result = "**"
    elif p_value <= 0.1:
        result = "*"
    else:
        result = " "
        
    return result
```

```python
test_COMBINED_DF = COMBINED_DF[ (COMBINED_DF.ANO_x ==2020) & (COMBINED_DF.MAGNITUD_x ==9) ].drop(columns = ['latitude',
                                                                                                             'longitude',
                                                                                                             'Precipitation'])
a = pg.pairwise_corr(data= test_COMBINED_DF, method ='spearman')
a['p-value-star'] = np.vectorize(assign_pvalue_star)(a['p-unc'])
sth = a[a.X == 'value_x'].round(2)
sth
```

```python
def combine(value, star):
    new_value = f"{value:.2f}"
    return new_value + " " + star
```

```python
years = [2019,2020,2022]
magnitude = [7,9,10,12]

year_frame_corr = []

for year in years:
    for mag in magnitude:
        test_COMBINED_DF = COMBINED_DF[ (COMBINED_DF.ANO_x ==year) & (COMBINED_DF.MAGNITUD_x ==mag) ].drop(columns = ['latitude',
                                                                                                             'longitude',
                                                                                                             'Precipitation'])
        a = pg.pairwise_corr(data= test_COMBINED_DF, method ='spearman')
        a['p-value-star'] = np.vectorize(assign_pvalue_star)(a['p-unc'])
        a['corr-with-star'] = np.vectorize(combine)(a['r'], a['p-value-star'])
        sth = a[a.X == 'value_x'].round(2)
        corr_pivot_table = sth.pivot_table(values ="r" , index = 'Y', columns = 'X' ).reset_index()
        corr_pivot_table['year'] = year
        corr_pivot_table['magnitude'] = mag
        year_frame_corr.append(corr_pivot_table)
```
